```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
import contextily as cx
```

```python
csv_file = "chess_players.csv"  # data compiled from FIDE ratings (https://ratings.fide.com/) and wikipedia biographies
```

```python
df = pd.read_csv(csv_file, sep=";", encoding="ISO-8859-1")
gdf = gpd.GeoDataFrame(df, geometry=gpd.points_from_xy(df.lon, df.lat), crs="epsg:4326")
gdf.head()
```

```python
f, ax = plt.subplots(figsize=(20, 20))
proj_str = "epsg:3857"  # WGS 84 / Pseudo-Mercator (https://epsg.io/3857)
gdf_proj = gdf.to_crs(proj_str)
minx, miny, maxx, maxy = gdf_proj.geometry.total_bounds
x_left, x_right, y_pad = 1100000, 700000, 1500000
ax.set_xlim(minx - x_left, maxx + x_right)
ax.set_ylim(miny - y_pad, maxy + y_pad)
gdf_proj.plot(
    ax=ax,
    column="rank",
    cmap="hot",
    markersize=100,
    edgecolor="black",
    linewidth=1.0,
    alpha=0.9,
    legend=True,
    legend_kwds={"bbox_to_anchor": (0.0276, 0.93), "fontsize": 7, "frameon": True},
    categorical=True,
)
cx.add_basemap(ax, crs=proj_str, source=cx.providers.Stamen.TonerLite, alpha=0.75)
ax.axis("off")
ax.margins(0)
ax.tick_params(left=False, labelleft=False, bottom=False, labelbottom=False)

plt.suptitle(
    "Place of birth of Chess Super Grandmasters (Nov 2022)",
    fontsize=30,
    fontname="Lucida Console",
    y=0.65,
    color="black",
    x=0.35,
)
plt.title(
    "Grandmasters with Elo rating from 2700 (white) to 2859 (black)",
    fontsize=17,
    fontname="Lucida Console",
    y=0.87,
    color="dimgrey",
    x=0.26,
)
plt.title(
    "Data source: FIDE ratings & wikipedia biographies",
    fontsize=8,
    color="black",
    y=0.00,
    loc="right",
)
plt.tight_layout()
plt.savefig("chess_nov2022.png", dpi=200, bbox_inches="tight", pad_inches=0)
```
