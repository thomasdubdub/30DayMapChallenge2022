```python
import numpy as np
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import contextily as cx
```

```python
# Data from https://cartographyvectors.com/map/1530-ukraine-with-regions
geojson_file = "ukraine-with-regions_1530.geojson"
```

```python
gdf = gpd.read_file(geojson_file)
```

```python
gdf.head()
```

```python
gdf["name"].unique() 
```

```python
# Data from UNESCO: https://www.unesco.org/en/articles/damaged-cultural-sites-ukraine-verified-unesco
unesco_map = {
    "Chernihiv Oblast": 15,
    "Kiev Oblast": 31,
    "Kharkiv Oblast": 51,
    "Zaporizhia Oblast": 7,
    "Zhytomyr Oblast": 2,
    "Donetsk Oblast": 60,
    "Luhansk Oblast": 62,
    "Sumy Oblast": 8,
    "Mykolaiv Oblast": 6,
    "Vinnytsia Oblast": 2,
    "Odessa Oblast": 1,
    "Dnipropetrovsk Oblast": 1,
}
```

```python
def nb_unesco(name):
    if name in unesco_map.keys():
        return unesco_map[name]
    return 0
```

```python
gdf["nb_damaged_sites"] = gdf["name"].apply(nb_unesco)
```

```python
gdf.head()
```

```python
f, ax = plt.subplots(figsize=(10, 10))
gdf_proj = gdf.to_crs(epsg=3857)  # 6381)

colors = [(0.25, 0, 0, c) for c in np.linspace(0, 1, 100)]
cmapred = mcolors.LinearSegmentedColormap.from_list("mycmap", colors, N=100)

import matplotlib.pylab as pl
from matplotlib.colors import ListedColormap

# Choose colormap
cmap = pl.cm.copper_r  # inferno_r
# Get the colormap colors
my_cmap = cmap(np.arange(cmap.N))
# Set alpha
my_cmap[:, -1] = np.linspace(0, 0.7, cmap.N)
# Create new colormap
my_cmap = ListedColormap(my_cmap)

gdf_proj.plot(
    ax=ax,
    column="nb_damaged_sites",
    cmap=my_cmap,
    legend=True,
    legend_kwds={"shrink": 0.4}
)
cx.add_basemap(
    ax, crs=gdf_proj.crs.to_string(), source=cx.providers.Stamen.Terrain, alpha=0.8
)
ax.axis("off")
ax.margins(0)
ax.tick_params(left=False, labelleft=False, bottom=False, labelbottom=False)
plt.title(
    "Number of cultural sites damaged in Ukraine Oblasts",
    fontsize=15,
    fontname="Lucida Console",
    y=0.95,
    color="black",
    alpha=0.75,
    loc="left",
)
plt.title(
    "UNESCO preliminary damage assessment (31/10/22)",
    fontsize=9,
    fontname="Lucida Console",
    y=0.90,
    color="tab:brown",
    alpha=0.98,
    x=0.25,
)
plt.title(
    "https://github.com/thomasdubdub",
    fontsize=7,
    color="whitesmoke",
    alpha=0.9,
    y=0.00,
    loc="right",
)
plt.tight_layout()
plt.savefig("ukraine.png", dpi=200, bbox_inches="tight", pad_inches=0)
```
