```python
import pandas as pd
import geopandas as gpd
from cartopy import crs as ccrs
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import contextily as cx
```

```python
geojson_file = "reseau-cyclable.geojson"
```

```python
gdf = gpd.read_file(geojson_file)
gdf.head()
```

```python
gdf.plot()
```

```python
gdf.typologie_simple.unique()
```

```python
gdf_2154 = gdf.to_crs(epsg=2154)  # Lambert-93
```

```python
f, ax = plt.subplots(figsize=(10, 10))
gdf_2154.query("typologie_simple == 'Pistes cyclables'").plot(
    ax=ax,
    color="tab:blue",
)
gdf_2154.query("typologie_simple == 'Bandes cyclables'").plot(
    ax=ax,
    color="tab:green",
)
gdf_2154.query("typologie_simple == 'Couloirs de bus ouverts aux vélos'").plot(
    ax=ax,
    color="tab:orange",
)
gdf_2154.query(
    "typologie_simple == 'Autres itinéraires cyclables (ex : Aires piétonnes - Contre-sens cyclables)'"
).plot(ax=ax, color="tab:purple", alpha=0.6)
cx.add_basemap(
    ax,
    crs=gdf_2154.crs.to_string(),
    source=cx.providers.OpenStreetMap.HOT,
    # zoom=14,
    alpha=0.7,
)

ax.axis("off")
ax.margins(0)
ax.tick_params(left=False, labelleft=False, bottom=False, labelbottom=False)

plt.title(
    " Paris \n bicycle \n path \n Network",
    fontsize=25,
    fontname="Arial Black",
    color="brown",  # "darkgreen",
    x=0.009,
    y=0.70,
    loc="left",
)

plt.title(
    "https://github.com/thomasdubdub",
    fontsize=8,
    color="dimgrey",
    alpha=0.9,
    y=0.00,
    loc="right",
)
blue_patch = mpatches.Patch(color="tab:blue", label="Reserved cycle paths")
green_patch = mpatches.Patch(color="tab:green", label="Shared cycle paths")
orange_patch = mpatches.Patch(color="tab:orange", label="Bus lanes open to cycles")
purple_patch = mpatches.Patch(color="tab:purple", label="Other cycle paths")
plt.legend(
    handles=[blue_patch, green_patch, orange_patch, purple_patch],
    fontsize=11,
    facecolor="snow",  # "#d1ffbd",  # "#b7e1a1",  # "#96ae8d",
)
plt.tight_layout()
plt.savefig("paris_cycling.png", dpi=150, bbox_inches="tight", pad_inches=0)
```
