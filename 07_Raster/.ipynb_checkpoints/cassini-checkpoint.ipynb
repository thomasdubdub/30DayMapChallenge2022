```python
import pandas as pd
import geopandas as gpd
import rasterio
import matplotlib.pyplot as plt
from shapely.geometry import Point
import contextily as cx
```

```python
# Tif to be downloaded here: https://davidrumsey.georeferencer.com/maps/783efdfd-9caf-503a-8397-bea7c21c3348/
src = rasterio.open("toulouse.tif") 
src.crs
```

```python
points = [
    Point(152300, 5.4086 * 1e6),
    Point(152000, 5.40065 * 1e6),
    Point(162150, 5.40650 * 1e6),
    Point(159760, 5.4017 * 1e6),
    Point(159780, 5.4002 * 1e6),
    Point(166500, 5.4025 * 1e6),
]
gdf = gpd.GeoDataFrame(
    ["Airport", "Golf", "Train station", "Stadium", "Casino", "Cit√© de l'Espace"],
    geometry=points,
)
```

```python
from rasterio.plot import show

fig, ax = plt.subplots(figsize=(10, 10))

# transform rasterio plot to real world coords
extent = [src.bounds[0], src.bounds[2], src.bounds[1], src.bounds[3]]
ax = rasterio.plot.show(src, extent=extent, ax=ax, cmap="pink", alpha=0.7)
ax.set_xlim(150000, 170000)
ax.set_ylim(5.395 * 1e6, 5.415 * 1e6)

gdf.plot(ax=ax, color="brown", marker="x")
for x, y, label in zip(
    gdf.geometry.x,
    gdf.geometry.y,
    gdf[0],
):
    ax.annotate(
        label,
        xy=(x, y),
        xytext=(-20, 7),
        textcoords="offset points",
        size=25.0,
        fontname="Brush Script MT",
        color="brown",  
    )

plt.title(
    "Cassini Map (38 - Toulouse) revisited",
    fontsize=45,
    fontname="Brush Script MT",
    color="brown", 
    x=0.020,
    y=0.90,
    loc="left",
)
plt.title(
    "https://github.com/thomasdubdub",
    fontsize=8,
    color="black",
    alpha=0.9,
    y=0.00,
    loc="right",
)
ax.axis("off")
ax.margins(0)
ax.tick_params(left=False, labelleft=False, bottom=False, labelbottom=False)
plt.tight_layout()
plt.savefig("cassini.png", dpi=150, bbox_inches="tight", pad_inches=0)
```
