```python
import osmnx as ox
import matplotlib.pyplot as plt
import contextily as cx
```

```python
osm_filter = '["aeroway"~"spaceport|launch_complex|launchpad|landingpad|hangar|runway"]'
# osm_filter = '["aeroway"~"launch_complex"]'
# osm_filter = '["aeroway"=*]'
place = "Boca Chica Village, Texas"
place = "Boca Chica State Park"
# place = "SpaceX Launch Facility"
# place = "Cape Canaveral, Floride"
lat, lon = 25.98753, -97.18662
lat, lon = 28.47400, -80.57753  # Cap Canaveral
```

```python
G = ox.graph_from_place(
    place,
    simplify=False,
    retain_all=True,
    # truncate_by_edge=True,
    buffer_dist=2000,
    # custom_filter=osm_filter,
)  # , custom_filter=osm_filter)  # , dist=5000)  # , custom_filter=osm_filter)
gdf_nodes, gdf_edges = ox.graph_to_gdfs(G)
print(gdf_nodes.shape)
print(gdf_edges.shape)
gdf_edges.head()
```

```python
gdf_edges.name.unique()
```

```python
gdf_edges.plot()
```

```python
f, ax = plt.subplots(figsize=(10, 10))
gdf_3857 = gdf_edges.to_crs("epsg:3857")
minx, miny, maxx, maxy = gdf_3857.geometry.total_bounds
x_pad, y_pad = 800, 600
ax.set_xlim(minx - x_pad, maxx + x_pad)
ax.set_ylim(miny - y_pad, maxy + y_pad)
ax.set_xlim(-1.08161 * 1e7, -1.08145 * 1e7)
ax.set_ylim(2.99811 * 1e6, 2.9995 * 1e6)
gdf_3857.plot(ax=ax)
cx.add_basemap(ax, alpha=0.9, zoom=19, source=cx.providers.OpenStreetMap.Mapnik)
```
