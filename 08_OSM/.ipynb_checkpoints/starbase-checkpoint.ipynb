```python
import osmnx as ox
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
from shapely.geometry import Point
import contextily as cx
```

```python
# https://wiki.openstreetmap.org/wiki/Tag:aeroway%3Dspaceport
osm_spaceport_filter = (
    '["aeroway"~"spaceport|launch_complex|launchpad|landingpad|hangar|runway"]'
)
place = "Boca Chica State Park"
```

```python
def get_gdfs(place, dist=1000, filter=None):
    G = ox.graph_from_place(
        place,
        simplify=False,
        retain_all=True,
        truncate_by_edge=True,
        buffer_dist=dist,
        custom_filter=filter,
    )
    return ox.graph_to_gdfs(G)
```

```python
gdf_n, gdf_e = get_gdfs(place, filter=osm_spaceport_filter)
gdf_all_n, gdf_all_e = get_gdfs(place)
```

```python
print(gdf_n.shape, gdf_e.shape, gdf_all_n.shape, gdf_all_e.shape)
```

```python
pgdf_n, pgdf_e, pgdf_all_n, pgdf_all_e = (
    gdf_n.to_crs("epsg:3857"),
    gdf_e.to_crs("epsg:3857"),
    gdf_all_n.to_crs("epsg:3857"),
    gdf_all_e.to_crs("epsg:3857"),
)
firsts = pgdf_e.groupby("name").first() # only one label for each polygon
firsts["geom"] = firsts["geometry"].apply(lambda x: Point(x.coords[0])) # first point of linestring
```

```python
f, ax = plt.subplots(figsize=(10, 10))
pgdf_all_e.plot(ax=ax, color="grey")
pgdf_e.plot(ax=ax, color="darkorange", linewidth=2.5)
for x, y, label in zip(
    firsts.geom.x,
    firsts.geom.y,
    firsts.index,
):
    ax.annotate(
        label,
        xy=(x, y),
        xytext=(-30, 15),
        textcoords="offset points",
        size=9.0,
        fontname="Lucida Console",
        weight="bold",
        color="darkorange",
    )

ax.set_xlim(-1.08157 * 1e7, -1.081490 * 1e7)
ax.set_ylim(2.9984 * 1e6, 2.99905 * 1e6)
cx.add_basemap(
    ax, alpha=0.75, zoom=18, source=cx.providers.CartoDB.DarkMatter
)
plt.title(
    'SpaceX "Starbase" spaceport at Boca Chica, Texas',
    fontsize=24,
    color="darkorange",
    weight="bold",
    x=0.017,
    y=0.90,
    loc="left",
)
plt.title(
    "https://github.com/thomasdubdub",
    fontsize=8,
    color="black",
    alpha=0.9,
    y=0.00,
    loc="right",
)
ax.axis("off")
ax.margins(0)
ax.tick_params(left=False, labelleft=False, bottom=False, labelbottom=False)
plt.tight_layout()
plt.savefig("starbase.png", dpi=150, bbox_inches="tight", pad_inches=0)
```
