```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
stable_url = (
    "https://www.data.gouv.fr/fr/datasets/r/eb36371a-761d-44a8-93ec-3d728bec17ce"
) # https://www.data.gouv.fr/fr/datasets/contours-des-departements-francais-issus-d-openstreetmap/
```

```python
gdf = gpd.read_file(stable_url)
```

```python
gdf.dtypes
```

```python
gdf.head()
```

```python
def get_value(code):
    if (code == "2A") or (code == "2B"):
        return 20
    if (code == "69D") or (code == "69M"):
        return 69
    return int(code)
```

```python
def get_color(value):
    if value >= 65:
        return "snow"
    return "dimgrey"
```

```python
gdf["value"] = gdf["code_insee"].apply(lambda x: get_value(x))
gdf["color"] = gdf["value"].apply(lambda x: get_color(x))
```

```python
gdf.shape
```

```python
exclude_list = {"971", "972", "973", "974", "976"}
```

```python
gdf.query("code_insee not in @exclude_list").plot(column="value", cmap="Reds")
```

```python
f, ax = plt.subplots(figsize=(10, 10))
gdf_proj = gdf.query("code_insee not in @exclude_list").to_crs("epsg:2154")
gdf_proj["center"] = gdf_proj.geometry.centroid
gdf_proj.plot(
    ax=ax,
    column="value",
    cmap="YlGn",  # "Reds",
    edgecolor="dimgrey",
)
for x, y, label, color in zip(
    gdf_proj.center.x,
    gdf_proj.center.y,
    gdf_proj["code_insee"],
    gdf_proj["color"],
):
    ax.annotate(
        label,
        xy=(x, y),
        xytext=(-2, 2),
        textcoords="offset points",
        size=10.0,
        color=color,
    )
plt.title(
    "French Departments coloring based on...\n their administrative code",
    fontsize=30,
    fontname="Consolas",
    color="darkgreen",
    y=1.05,
    loc="center",
)
plt.title(
    "https://github.com/thomasdubdub",
    fontsize=8,
    color="black",
    alpha=0.9,
    y=0.00,
    loc="left",
)
ax.axis("off")
ax.margins(0)
ax.tick_params(left=False, labelleft=False, bottom=False, labelbottom=False)
plt.tight_layout()
plt.savefig("bad_depts.png", dpi=150, bbox_inches="tight", pad_inches=0)
```
