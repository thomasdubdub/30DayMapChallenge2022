```python
import osmnx as ox
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
from shapely.geometry import Point
import contextily as cx
```

```python
place = "Red Light District Amsterdam"
custom_filter = '["tourism"~"attraction"]'
```

```python
def get_gdfs(place, dist=1000, custom_filter=None):
    G = ox.graph_from_place(
        place,
        simplify=False,
        retain_all=True,
        truncate_by_edge=True,
        buffer_dist=dist,
        custom_filter=custom_filter,
    )
    return ox.graph_to_gdfs(G)
```

```python
gdf_n, gdf_e = get_gdfs(place, dist=0.1, custom_filter=custom_filter)
gdf_all_n, gdf_all_e = get_gdfs(place)
```

```python
pgdf_n, pgdf_e, pgdf_all_n, pgdf_all_e = (
    gdf_n.to_crs("epsg:3857"),
    gdf_e.to_crs("epsg:3857"),
    gdf_all_n.to_crs("epsg:3857"),
    gdf_all_e.to_crs("epsg:3857"),
)
```

```python
plt.rcParams["axes.facecolor"] = "black"
plt.rcParams["savefig.facecolor"] = "black"
plt.rcParams["figure.facecolor"] = "black"
f, ax = plt.subplots(figsize=(10, 10))
pgdf_e.plot(ax=ax, color="red", linewidth=3.0)
pgdf_all_e.plot(ax=ax, color="grey", alpha=0.5)

plt.title(
    " The \n Red \n Light \n District",
    fontsize=25,
    fontname="Arial Black",
    color="red",
    y=0.83,
    loc="left",
)

plt.title(
    "https://github.com/thomasdubdub",
    fontsize=8,
    color="grey",
    alpha=0.9,
    y=0.00,
    loc="right",
)
ax.axis("off")
ax.margins(0)
ax.tick_params(left=False, labelleft=False, bottom=False, labelbottom=False)
plt.tight_layout()
plt.savefig("red.png", dpi=150, bbox_inches="tight", pad_inches=0)
```
