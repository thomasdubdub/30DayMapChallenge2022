```python
import osmnx as ox
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
from matplotlib_scalebar.scalebar import ScaleBar
from shapely.geometry import Point
import contextily as cx
```

```python
place = "Principality of Monaco"
custom_filter = '["tourism"~"attraction"]'
```

```python
def get_gdfs(place, dist=1000, custom_filter=None):
    G = ox.graph_from_place(
        place,
        simplify=False,
        retain_all=True,
        truncate_by_edge=True,
        buffer_dist=dist,
        custom_filter=custom_filter,
    )
    return ox.graph_to_gdfs(G)
```

```python
gdf_n, gdf_e = get_gdfs(place, dist=0)
gdf_all_n, gdf_all_e = get_gdfs(place)
```

```python
gdf_e.plot()
```

```python
pgdf_n, pgdf_e = gdf_n.to_crs("epsg:3857"), gdf_e.to_crs("epsg:3857")
```

```python
plt.rcParams["axes.facecolor"] = "black"
plt.rcParams["savefig.facecolor"] = "black"
plt.rcParams["figure.facecolor"] = "black"
f, ax = plt.subplots(figsize=(10, 10))
pgdf_e.plot(ax=ax, color="snow", linewidth=1.0)
plt.title(
    "The Principality",
    fontsize=45,
    fontname="Constantia",
    color="red",
    y=0.90,
    x=0.06,
    loc="left",
)
plt.title(
    "of Monaco",
    fontsize=37,
    fontname="Constantia",
    color="white",
    y=0.83,
    x=0.33,
)
ax.add_artist(
    ScaleBar(
        2.0,
        location="lower right",
        color="red",
        box_color="black",
        box_alpha=0.9,
        font_properties={"weight": "bold"},
    )
)
ax.axis("off")
ax.margins(0)
ax.tick_params(left=False, labelleft=False, bottom=False, labelbottom=False)
plt.tight_layout()
plt.savefig("monaco.png", dpi=150, bbox_inches="tight", pad_inches=0)
```
