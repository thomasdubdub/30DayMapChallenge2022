```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
url = "https://www.data.gouv.fr/fr/datasets/r/64e02cff-9e53-4cb2-adfd-5fcc88b2dc09"
```

```python
df = pd.read_csv(url, sep=";")
df.head()
```

```python
sp95 = df.query("prix_nom == 'SP95'")
sp95["lat"] = sp95["geom"].apply(lambda x: float(x.split(",")[0]))
sp95["lon"] = sp95["geom"].apply(lambda x: float(x.split(",")[1]))
sp95.head()
```

```python
gdf = gpd.GeoDataFrame(
    sp95, geometry=gpd.points_from_xy(sp95["lon"], sp95["lat"]), crs="epsg:4326"
)
gdf.head()
```

```python
gdf.plot()
```

```python
f, ax = plt.subplots(figsize=(10, 8))
gdf.plot(
    ax=ax,
    kind="hexbin",
    x="lon",
    y="lat",
    C="prix_valeur",
    gridsize=(20, 20),
    cmap="Reds",
)

plt.title(
    "Price of SP95 petrol in France, 14/11/22",
    fontsize=25,
    fontname="Consolas",
    color="brown",
    y=1.0,
)

plt.title(
    "https://github.com/thomasdubdub",
    fontsize=8,
    color="brown",
    alpha=0.9,
    y=0.0,
    loc="left",
)
ax.axis("off")
ax.tick_params(left=False, labelleft=False, bottom=False, labelbottom=False)
plt.tight_layout()
plt.savefig("sp95.png", dpi=150, bbox_inches="tight", pad_inches=0)
```

```python
stable_url = (
    "https://www.data.gouv.fr/fr/datasets/r/eb36371a-761d-44a8-93ec-3d728bec17ce"
)
deps = gpd.read_file(stable_url)
deps.head()
```

```python
def get_value(code):
    if code == "69D" or code == "69M":
        return s["69"]
    return s[code]
```

```python
exclude_list = ["971", "972", "973", "974", "975", "976"]
fdeps = deps.query("code_insee not in @exclude_list")
fdeps["C"] = fdeps["code_insee"].apply(lambda x: get_value(x))
```

```python
fdeps["center"] = fdeps["geometry"].centroid
fdeps["x"] = fdeps["center"].x
fdeps["y"] = fdeps["center"].y
fdeps.head()
```

```python
fdeps.plot(kind="hexbin", x="x", y="y", C="C", gridsize=(10, 10))
```
