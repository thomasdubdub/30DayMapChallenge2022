```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
import numpy as np
import contextily as cx
import matplotlib.patches as mpatches
```

```python
url = "https://www.data.gouv.fr/fr/datasets/r/38276359-ca66-4566-b4c5-f012286a77b4"
# Open data OSM here: https://www.data.gouv.fr/fr/datasets/decoupage-administratif-communal-francais-issu-d-openstreetmap/
osm_fp = "communes-20220101-shp/communes-20220101.shp"
```

```python
com_gdf = gpd.read_file(osm_fp)
com_gdf.head()
```

```python
insee_geometry_dict = dict(zip(com_gdf.insee, com_gdf.geometry))
```

```python
df = pd.read_csv(url, sep=";", encoding="ISO-8859-1")
df.head()
```

```python
cow_aop = [
    "Bleu d'Auvergne",
    "Cantal",
    "Fourme d'Ambert",
    "Saint-Nectaire",
    "Salers",
    "Bleu du Vercors-Sassenage",
    "Fourme de Montbrison",
    "Abondance",
    "Beaufort",
    "Reblochon de Savoie",  # "Reblochon",
    "Tome des Bauges",
    "Bleu de Gex haut Jura ou Bleu de Septmoncel",  # Bleu de Gex Haut-Jura
    "Chaource",
    "Comté",
    "Epoisses",  # "Époisses",
    "Mont d'Or ou Vacherin du Haut-Doubs",  # "Mont d'Or",
    "Morbier",
    "Bleu des Causses",
    "Laguiole",
    "Camembert de Normandie",
    "Livarot",
    "Neufchâtel",
    "Pont-l'Évêque",
    "Maroilles",
    "Brie de Melun",
    "Brie de Meaux",
    "Langres",
    "Munster",
]
goat_aop = [
    "Picodon",
    "Rigotte de Condrieu",
    "Chevrotin",
    "Charolais",
    "Mâconnais",
    "Banon",
    "Brousse du Rove",
    "Brocciu",
    "Pélardon",
    "Rocamadour",
    "Chabichou du Poitou",
    "Chavignol",  # "Crottin de Chavignol",
    "Pouligny-Saint-Pierre",  # "Pouligny Saint-Pierre"
    "Sainte-Maure de Touraine",
    "Selles-sur-Cher",
    "Valençay",
]
sheep_aop = [
    "Brocciu",
    "Roquefort",
    "Ossau-Iraty",
]
print(len(cow_aop), len(goat_aop), len(sheep_aop))
```

```python
def get_geom(insee):
    if insee in insee_geometry_dict.keys():
        return insee_geometry_dict[insee]
    return np.NaN
```

```python
def get_gdf(cheese_names_list):
    sdf = df[df["Aire géographique"].isin(cheese_names_list)]
    sdf = sdf.drop_duplicates(subset=["CI"])
    sdf["geometry"] = sdf["CI"].apply(lambda x: get_geom(x))
    sdf = sdf[sdf["geometry"].notna()]
    return gpd.GeoDataFrame(sdf, geometry="geometry", crs="epsg:4326")
```

```python
cow_gdf = get_gdf(cow_aop).to_crs("epsg:2154")
goat_gdf = get_gdf(goat_aop).to_crs("epsg:2154")
sheep_gdf = get_gdf(sheep_aop).to_crs("epsg:2154")
```

```python
f, ax = plt.subplots(figsize=(10, 10))
ax.set_xlim(0.1 * 1e6, 1.3 * 1e6)
ax.set_ylim(6.0 * 1e6, 7.2 * 1e6)
cow_gdf.plot(ax=ax, color="tab:red", alpha=0.5, edgecolor="snow", linewidth=0.2)
goat_gdf.plot(ax=ax, color="tab:blue", alpha=0.5, edgecolor="snow", linewidth=0.2)
sheep_gdf.plot(ax=ax, color="yellow", alpha=0.5, edgecolor="snow", linewidth=0.2)
cx.add_basemap(ax, crs="epsg:2154", source=cx.providers.Stamen.TonerLite, alpha=0.75)
plt.title(
    "French AOP Cheese areas",
    fontsize=45,
    fontname="Consolas",
    color="black",
    y=0.93,
    loc="left",
)
plt.title(
    "https://github.com/thomasdubdub",
    fontsize=8,
    color="black",
    alpha=0.9,
    y=0.00,
    loc="right",
)

ax.axis("off")
ax.margins(0)
ax.tick_params(left=False, labelleft=False, bottom=False, labelbottom=False)
plt.tight_layout()
patch1 = mpatches.Patch(color="tab:red", label="Cow cheese")
patch2 = mpatches.Patch(color="tab:blue", label="Goat cheese")
patch3 = mpatches.Patch(color="yellow", label="Sheep cheese")
plt.legend(handles=[patch1, patch2, patch3])

plt.savefig("cheese.png", dpi=150, bbox_inches="tight", pad_inches=0)
```
