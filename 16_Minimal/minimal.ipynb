```python
import matplotlib.pyplot as plt
import pandas as pd
import geopandas as gpd
from cartopy import crs as ccrs
import contextily as cx
```

```python
path = gpd.datasets.get_path("naturalearth_lowres")
df = gpd.read_file(path)
```

```python
crs = ccrs.Mollweide()  # AzimuthalEquidistant()
crs_proj4 = crs.proj4_init
df_ae = df.to_crs(crs_proj4)
df_ae.plot()
```

```python
color = "ghostwhite"
fig = plt.figure(figsize=(10, 10))
ax = plt.axes(projection=crs)
df_ae.plot(ax=ax, color=color, alpha=1.0)
ax.outline_patch.set_edgecolor(color)
ax.outline_patch.set_alpha(0.4)
plt.savefig("minimal.png", dpi=150, bbox_inches="tight", pad_inches=0)
```
