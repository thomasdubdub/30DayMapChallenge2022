```python
import pandas as pd
import geopandas as gpd
import contextily as cx
from shapely.geometry import Point
import matplotlib.pyplot as plt
```

```python
points = [
    Point(108.08638888888889, 53.249722222222225),
]
gdf = gpd.GeoDataFrame(["Max depth"], geometry=points, crs="epsg:4326")
```

```python
gdf.plot()
```

```python
crs = "epsg:3857"
gdf_proj = gdf.to_crs(crs)
```

```python
blue_color = "skyblue"
f, ax = plt.subplots(figsize=(10, 10))
gdf_proj.plot(ax=ax, color="navy", marker="x", markersize=20)

ax.set_xlim(1.14 * 1e7, 1.25 * 1e7)
ax.set_ylim(6.6 * 1e6, 7.6 * 1e6)

for x, y, label in zip(
    gdf_proj.geometry.x,
    gdf_proj.geometry.y,
    gdf_proj[0],
):
    ax.annotate(
        label,
        xy=(x, y),
        xytext=(-35, 10),
        textcoords="offset points",
        size=20.0,
        color="navy",
    )
cx.add_basemap(
    ax,
    crs=crs,
    source=cx.providers.Stamen.Watercolor,
    alpha=0.9,
)

ax.axis("off")
ax.margins(0)
ax.tick_params(left=False, labelleft=False, bottom=False, labelbottom=False)

plt.title(
    " Deepest \n lake \n in the world",
    fontsize=50,
    fontname="Palatino Linotype",
    color="navy",
    x=0.11,
    y=0.68,
    loc="left",
)

plt.title(
    "https://github.com/thomasdubdub",
    fontsize=8,
    color="k",
    alpha=0.7,
    y=0.00,
    loc="right",
)

plt.tight_layout()
plt.savefig("bluelake.png", dpi=150, bbox_inches="tight", pad_inches=0)
```
