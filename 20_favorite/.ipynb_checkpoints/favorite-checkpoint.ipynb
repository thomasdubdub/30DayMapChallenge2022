```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
from shapely.geometry import Point, LineString
import contextily as cx
```

```python
csv_file = "tour.csv"
logo_file = "logo.png"
```

```python
df = pd.read_csv(
    csv_file,
    sep=";",
    encoding="ISO-8859-1",
    names=["date", "place", "lat", "lon"],
    header=None,
)
gdf = gpd.GeoDataFrame(df, geometry=gpd.points_from_xy(df.lon, df.lat), crs="epsg:4326")
gdf.head()
```

```python
key_dates = [
    "14/01/2023",
    "25/02/2023",
    "02/03/2023",
    "02/04/2023",
    "27/05/2023",
    "03/06/2023",
    "01/07/2023",
]
```

```python
f, ax = plt.subplots(figsize=(10, 6))
ax.set_xlim(-1.4 * 1e7, 0.25 * 1e7)
ax.set_ylim(1 * 1e6, 9 * 1e6)
proj_str = "epsg:3857"  # WGS 84 / Pseudo-Mercator (https://epsg.io/3857)
gdf_proj = gdf.to_crs(proj_str)
gdf_proj.plot(ax=ax, color="yellow", markersize=5, alpha=0.8)
cx.add_basemap(
    ax, crs=proj_str, source=cx.providers.CartoDB.DarkMatterNoLabels, alpha=0.7
)
nb_color = "white"
date_color = "grey"
for x, y, label in zip(
    gdf_proj.geometry.x,
    gdf_proj.geometry.y,
    gdf_proj.index,
):
    ax.annotate(
        label,
        xy=(x, y),
        xytext=(-2.5, 5),
        textcoords="offset points",
        size=7.0,
        color=nb_color,
    )
s_gdf_proj = gdf_proj.query("date in @key_dates")
for x, y, label in zip(
    s_gdf_proj.geometry.x,
    s_gdf_proj.geometry.y,
    s_gdf_proj.date,
):
    ax.annotate(
        label,
        xy=(x, y),
        xytext=(-15.0, -10),
        textcoords="offset points",
        size=7.0,
        color=date_color,
    )

gdf_proj["LINE"] = [
    (LineString([[a.x, a.y], [b.x, b.y]]) if b != None else None)
    for (a, b) in zip(gdf_proj.geometry, gdf_proj.geometry.shift(-1, axis=0))
]
gdf2_proj = gdf_proj.copy()
gdf2_proj["geometry"] = gdf2_proj["LINE"]
gdf2_proj.plot(ax=ax, color="red", linewidth=1.0, alpha=0.8)
plt.title(
    "Will of the People World Tour 2023",
    fontsize=28,
    fontname="Lucida Console",  # "Consolas",
    fontweight="bold",
    color="k",
    x=0.1,
    y=0.88,
    loc="left",
)
plt.title(
    "https://github.com/thomasdubdub",
    fontsize=8,
    color="grey",
    alpha=0.9,
    y=0.00,
    loc="right",
)
im = plt.imread(logo_file)
newax = f.add_axes([0.73, 0.1, 0.25, 0.25], anchor="NW", zorder=1)
newax.imshow(im)
newax.axis("off")
ax.axis("off")
ax.margins(0)
ax.tick_params(left=False, labelleft=False, bottom=False, labelbottom=False)
plt.tight_layout()
plt.savefig("muse.png", dpi=200, bbox_inches="tight", pad_inches=0)
```
