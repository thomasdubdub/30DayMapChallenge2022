```python
import pandas as pd
import geopandas as gpd
from shapely.geometry import Point
import matplotlib.pyplot as plt
import contextily as cx
```

```python
points = [
    Point(0.0, 0.0),
]
gdf = gpd.GeoDataFrame(["Max depth"], geometry=points, crs="epsg:4326")
crs = "epsg:3857"
gdf_proj = gdf.to_crs(crs)
```

```python
plt.rcParams["axes.facecolor"] = "black"
plt.rcParams["savefig.facecolor"] = "black"
plt.rcParams["figure.facecolor"] = "black"
f, ax = plt.subplots(figsize=(10, 10))
gdf_proj.plot(ax=ax, color="navy", marker="x", markersize=20)
ax.set_xlim(-0.3 * 1e4, 0.3 * 1e4)
ax.set_ylim(-0.3 * 1e4, 0.3 * 1e4)
cx.add_basemap(ax, crs=crs, source=cx.providers.Strava.Run, alpha=0.9, zoom=11)
ax.axis("off")
ax.margins(0)
ax.tick_params(left=False, labelleft=False, bottom=False, labelbottom=False)
plt.title(
    "Running on Null Island",
    fontsize=40,
    fontname="Lucida Console",
    color="white",
    y=0.92,
    loc="center",
)
plt.title(
    "Strava recording of running activity on 0°N 0°E Island",
    fontsize=15,
    fontname="Lucida Console",
    color="white",
    y=0.88,
    loc="left",
    x=0.15,
)
plt.title(
    "https://github.com/thomasdubdub",
    fontsize=8,
    color="snow",
    alpha=0.7,
    y=0.00,
    loc="right",
)
plt.tight_layout()
plt.savefig("null.png", dpi=150, bbox_inches="tight", pad_inches=0)
```
