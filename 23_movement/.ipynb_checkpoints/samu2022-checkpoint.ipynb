```python
import pandas as pd
import geopandas as gpd
from traffic.core import Traffic
import glob
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import contextily as cx
from distinctipy import distinctipy
```

```python
samu_folder = "samu/"
```

```python
def traffic_to_gdf(t):
    df = t.data
    df.dropna(subset=["callsign", "latitude", "longitude"], inplace=True)
    return gpd.GeoDataFrame(
        df,
        geometry=gpd.points_from_xy(df["longitude"], df["latitude"]),
        crs="epsg:4326",
    )
```

```python
gdf_list = []
for f in glob.glob(samu_folder + "*.pkl"):
    print(f)
    t = Traffic.from_file(f)
    gdf_list.append(traffic_to_gdf(t))

fgdf = gpd.GeoDataFrame(pd.concat(gdf_list, ignore_index=True), crs=gdf_list[0].crs)
print(fgdf.shape)
```

```python
callsigns = sorted(list(fgdf.callsign.unique()))
nb_colors = len(callsigns)
print(nb_colors)
# generate N visually distinct colours
colors = distinctipy.get_colors(nb_colors)
color_mapping = {callsign: color for callsign, color in zip(callsigns, colors)}
```

```python
f, ax = plt.subplots(figsize=(10, 10))
fgdf.to_crs("epsg:3857").plot(
    ax=ax, markersize=0.002, alpha=0.5, color=fgdf["callsign"].map(color_mapping)
)
ax.set_xlim(-0.5 * 1e6, 0.9 * 1e6)
ax.set_ylim(5.25 * 1e6, 6.65 * 1e6)

cx.add_basemap(ax, source=cx.providers.Stamen.TonerLite, alpha=0.6)

plt.title(
    "Oprations SAMU en 2022 (ADS-B OpenSky Network)",
    fontsize=20,
    fontname="Consolas",
    color="k",
    y=1.0,
    loc="center",
)
plt.title(
    "https://github.com/thomasdubdub",
    fontsize=8,
    color="black",
    alpha=0.9,
    y=0.00,
    loc="right",
)

patches = [
    mpatches.Patch(color=color_mapping[x], label=x) for x in sorted(color_mapping)
]
plt.legend(handles=patches, prop={"size": 8}, bbox_to_anchor=(1.15, 1.0))
ax.axis("off")
ax.margins(0)
ax.tick_params(left=False, labelleft=False, bottom=False, labelbottom=False)
plt.tight_layout()
plt.savefig("samu2022.png", dpi=300, bbox_inches="tight", pad_inches=0)
```
