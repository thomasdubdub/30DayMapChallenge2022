```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
import contextily as cx
import matplotlib.patches as mpatches
from cartopy import crs as ccrs
```

```python
path = gpd.datasets.get_path("naturalearth_lowres")
gdf = gpd.read_file(path)
```

```python
gdf.head()
```

```python
oceania_continent_list = ["Oceania", "North America", "South America"]
oceania_country_list = [
    "Iceland",
    "United Kingdom",
    "Ireland",
    "Madagascar",
    "Mozambique",
    "Zimbabwe",
    "Botswana",
    "Namibia",
    "South Africa",
    "Zambia",
    "Angola",
]
asia_country_list = [
    "China",
    "Mongolia",
    "Turkmenistan",
    "Afghanistan",
    "Iran",
    "Nepal",
    "Uzbekistan",
    "Mongolia",
    "Bhutan",
    "Tajikistan",
    "Kyrgyzstan",
    "Turkmenistan",
    "Pakistan",
]
european_country_list = ["Kazakhstan", "Turkey", "Armenia", "Azerbaijan", "Georgia"]
```

```python
crs = ccrs.Mollweide()
crs_proj4 = crs.proj4_init
gdf_proj = gdf.to_crs(crs_proj4)
```

```python
bckcolor = "powderblue"
lw = 0.9
plt.rcParams["axes.facecolor"] = bckcolor
plt.rcParams["savefig.facecolor"] = bckcolor
plt.rcParams["figure.facecolor"] = bckcolor
f, ax = plt.subplots(figsize=(15, 10))

gdf_proj.plot(ax=ax, color="snow", edgecolor="dimgrey", alpha=0.99, linewidth=lw)
gdf_proj.query("continent in @oceania_continent_list").plot(ax=ax, color="tab:blue")
gdf_proj.query("continent == 'Europe'").plot(ax=ax, color="tab:orange")
gdf_proj.query("name in @european_country_list").plot(
    ax=ax, color="tab:orange", alpha=1.0
)
gdf_proj.query("name in @oceania_country_list").plot(ax=ax, color="tab:blue", alpha=1.0)
gdf_proj.query("name in @asia_country_list").plot(ax=ax, color="tab:green", alpha=1.0)

patch1 = mpatches.Patch(color="tab:blue", label="Oceania")
patch2 = mpatches.Patch(color="tab:orange", label="Eurasia")
patch3 = mpatches.Patch(color="tab:green", label="Eastasia")
patch4 = mpatches.Patch(color="snow", label="Disputed")
plt.legend(handles=[patch1, patch2, patch3, patch4], prop={"size": 18})
plt.title(
    " 1984 \n in \n 2022",
    fontsize=55,
    fontname="Consolas",
    color="black",
    y=0.45,
    loc="left",
)
plt.title(
    "https://github.com/thomasdubdub",
    fontsize=8,
    color="black",
    alpha=0.9,
    y=0.00,
    loc="right",
)

ax.axis("off")
ax.margins(0)
ax.tick_params(left=False, labelleft=False, bottom=False, labelbottom=False)
plt.tight_layout()
plt.savefig("1984.png", dpi=150, bbox_inches="tight", pad_inches=0)
```
