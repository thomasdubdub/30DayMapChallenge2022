```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
import contextily as cx
import matplotlib.patches as mpatches
```

```python
# https://www.data.gouv.fr/fr/datasets/indices-de-position-sociale-geolocalises-des-ecoles-et-colleges-de-france-metropolitaine-et-des-drom-2/
csv_file = "ips-all-geoloc.csv" 
dom = ["971", "972", "973", "974", "976"]
```

```python
df = pd.read_csv(csv_file)
gdf = gpd.GeoDataFrame(df, geometry=gpd.points_from_xy(df.lon, df.lat), crs="epsg:4326")
gdf["type"] = gdf["nom_etab"].apply(lambda x: x[:7])
gdf.sample(5)
```

```python
f, ax = plt.subplots(figsize=(10, 10))
gdf_plot = (
    gdf.query("type == 'COLLEGE'").query("code_dept not in @dom").to_crs("epsg:3857")
)
print(gdf_plot.shape)
gdf_plot.query("secteur == 'public'").plot(
    ax=ax, color="tab:orange", markersize=5, alpha=0.6
)
print(gdf_plot.query("secteur == 'public'").shape)
print(gdf_plot.query("secteur != 'public'").shape)
gdf_plot.query("secteur != 'public'").plot(
    ax=ax, color="mediumblue", markersize=5, alpha=0.6
)
cx.add_basemap(ax, source=cx.providers.Stamen.TonerLines, alpha=0.6)
patch1 = mpatches.Patch(color="tab:orange", label="Public")
patch2 = mpatches.Patch(color="mediumblue", label="Private")
plt.legend(handles=[patch1, patch2], prop={"size": 15})
plt.title(
    "High schools in France",
    fontsize=41,
    color="dimgray",
    y=0.95,
    loc="left",
    x=0.12,
)
plt.title(
    "https://github.com/thomasdubdub",
    fontsize=8,
    color="k",
    alpha=0.9,
    y=0.00,
    loc="right",
)
ax.axis("off")
ax.margins(0)
ax.tick_params(left=False, labelleft=False, bottom=False, labelbottom=False)
plt.tight_layout()
plt.savefig("two.png", dpi=150, bbox_inches="tight", pad_inches=0)
```
