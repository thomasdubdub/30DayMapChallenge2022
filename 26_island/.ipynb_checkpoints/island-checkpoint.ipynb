```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
import rasterio
from rasterio.plot import show
import earthpy.plot as ep
import earthpy.spatial as es
import numpy as np
```

```python
# Download tiff file here: https://www.data.gouv.fr/fr/datasets/cartes-topographiques-de-tahiti/
tif_file = "mnt-idv-tht-2013-0.tiff"
```

```python
src = rasterio.open(tif_file)
```

```python
show(src)
```

```python
elevation = src.read(1)
hillshade = es.hillshade(elevation)
hillshade_azimuth_210 = es.hillshade(elevation, azimuth=210)
top_right = hillshade_azimuth_210[0, -1]
hillshade_azimuth_210[hillshade_azimuth_210 == top_right] = np.nan
```

```python
ep.plot_bands(
    elevation,
    cmap="gist_earth",
    title="DTM Without Hillshade",
    vmax=3000,
    figsize=(10, 6),
)
```

```python
plt.rcParams["axes.facecolor"] = "white"
plt.rcParams["savefig.facecolor"] = "white"
plt.rcParams["figure.facecolor"] = "white"
fig, ax = plt.subplots(1, figsize=(12, 12))
ep.plot_bands(
    elevation,
    ax=ax,
    cmap="terrain",
    cbar=False,
    vmin=-750,
    vmax=2800,
)
ax.imshow(hillshade_azimuth_210, cmap="Greys", alpha=0.5)

plt.title(
    "Tahiti Island (Lidar DEM)",
    fontsize=40,
    fontname="Consolas",
    color="seagreen",
    y=0.96,
    loc="center",
)
plt.title(
    "https://github.com/thomasdubdub",
    fontsize=8,
    color="black",
    alpha=0.9,
    y=0.00,
    loc="left",
)
ax.axis("off")
ax.margins(0)
ax.tick_params(left=False, labelleft=False, bottom=False, labelbottom=False)
plt.tight_layout()
plt.savefig("island.png", dpi=150, bbox_inches="tight", pad_inches=0)
plt.show()
```
