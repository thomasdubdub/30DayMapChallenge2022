```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
from matplotlib.offsetbox import OffsetImage, AnnotationBbox
import contextily as cx
```

```python
choirs_url = "choirs.csv"
osm_fp = "communes-20220101-shp/communes-20220101.shp"
laposte_file = "laposte_hexasmal.csv"
```

```python
laposte_df = pd.read_csv(laposte_file, sep=";")
laposte_df.head()
```

```python
codepostal_insee = dict(zip(laposte_df.code_postal, laposte_df.code_commune_insee))
```

```python
df = pd.read_csv(
    choirs_url,
    encoding="ISO-8859-1",
    sep=";",
    header=None,
    names=["name", "codepostal", "shortname"],
)
df.head()
```

```python
df["insee"] = df["codepostal"].apply(lambda x: codepostal_insee[int(x)])
```

```python
com_gdf = gpd.read_file(osm_fp)
com_gdf["center"] = com_gdf["geometry"].to_crs("epsg:3857").centroid
com_gdf.head()
```

```python
insee_center = dict(zip(com_gdf.insee, com_gdf.center))
```

```python
def get_geom(insee_code):
    if insee_code[:2] == "75":
        return insee_center["75056"]
    if insee_code[:2] == "69":
        return insee_center["69123"]
    if insee_code[:2] == "13":
        return insee_center["13055"]
    return insee_center[insee_code]
```

```python
df["geometry"] = df["insee"].apply(lambda x: get_geom(x))
gdf = gpd.GeoDataFrame(df, geometry="geometry", crs="epsg:3857")
gdf.head()
```

```python
choir_icon_path = "musical-notes-64.png"
choir = plt.imread(choir_icon_path)
f, ax = plt.subplots(figsize=(10, 10))
gdf.plot(ax=ax, color="red", alpha=0.0)
for x, y, label in zip(
    gdf.geometry.x,
    gdf.geometry.y,
    gdf.shortname,
):
    ax.annotate(
        label,
        xy=(x, y),
        xytext=(-35, 10),
        textcoords="offset points",
        size=7.0,
        color="grey",
        alpha=0.8,
    )
    ax.add_artist(
        AnnotationBbox(
            OffsetImage(choir, zoom=0.4, alpha=0.8),
            (x, y),
            frameon=False,
        )
    )

ax.set_xlim(-500000, 1000000)
ax.set_ylim(5.2 * 1e6, 6.7 * 1e6)
cx.add_basemap(ax, source=cx.providers.CartoDB.Positron)
plt.title(
    "Choir schools in France",
    fontsize=40,
    fontname="Consolas",
    color="k",
    y=0.95,
    loc="center",
)
plt.title(
    "https://github.com/thomasdubdub",
    fontsize=8,
    color="black",
    alpha=0.9,
    y=0.00,
    loc="right",
)
ax.axis("off")
ax.margins(0)
ax.tick_params(left=False, labelleft=False, bottom=False, labelbottom=False)
plt.tight_layout()
plt.savefig("music.png", dpi=150, bbox_inches="tight", pad_inches=0)
```
